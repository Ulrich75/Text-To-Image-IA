```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")

# if using torch < 2.0
# pipe.enable_xformers_memory_efficient_attention()

prompt = "An astronaut riding a green horse"

images = pipe(prompt=prompt).images[0]
```

```python
# Convert the image tensor to a NumPy array (if necessary)
if isinstance(images, torch.Tensor):
  image_numpy = images.squeeze().cpu().numpy()
else:
  image_numpy = images  # Already a NumPy array

# No slicing needed for single image (remove the problematic line)

# Display the single image using matplotlib
plt.imshow(image_numpy)
plt.show()
```
